```
UMBC_FEE = dataframe2.loc[dataframe2['INSTNM'] == 'University of Maryland-Baltimore County', 'TUITIONFEE_IN'].iloc[0]
print(UMBC_FEE)
list1.append(UMBC_FEE)
```

```
UMCP_FEE = dataframe2.loc[dataframe2['INSTNM'] == 'University of Maryland-College Park', 'TUITIONFEE_IN'].iloc[0]
print(UMCP_FEE)
list1.append(UMCP_FEE)
```

```
LUM_FEE = dataframe2.loc[dataframe2['INSTNM'] == 'Loyola University Maryland', 'TUITIONFEE_IN'].iloc[0]
print(LUM_FEE)
list1.append(LUM_FEE)
```

```
JHU_FEE = dataframe2.loc[dataframe2['INSTNM'] == 'Johns Hopkins University', 'TUITIONFEE_IN'].iloc[0]
print(JHU_FEE)
list1.append(JHU_FEE)
```

```
print(list1)
```

```
min(list1)
```

```
max(list1)
```

```
sum = 0
mean = 0
for i in list1:
  sum+=i
mean = sum / len(list1)
print(mean)
```
