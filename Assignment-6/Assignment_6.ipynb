5. Write a function named cleanse_year() that takes a string as input and removes any asterisks in the string and return the cleansed string. Test the function by using test strings (for example, "1989*", "20*10")

```
def cleanse_year(cost):
    cost = int (cost.strip('*'))
    return cost
```

6. Cleanse the column "Census Year" using the function cleanse_year() and Pandas's apply() function.

```
data1["Census Year"] = data1["Census Year"].apply(cleanse_year)
data1
```

7. Repeat this process for column "Total Population", "Census Cost", "Average Cost Per Person". (define, test, and apply the function)

```
data1["Total Population"] = data1["Total Population"].astype(int)
```

```
def cleanse_cost(cost):
    n_cost = cost.lstrip("$").replace(",","")
    if  cost.find("4.5") != -1:          
        n_cost = "45"+"0" * 8
    elif cost.find("Billion") != -1:     
        n_cost = "13"+"0" * 9
    return int (n_cost)
```

* Cleanse the column "Census Cost" using the function cleanse_cost() and Pandas's apply() function.

```
data1["Census Cost"] = data1["Census Cost"].apply(cleanse_cost)
data1
```

```
def cleanse_cent(cost):
    if  cost.find("cent") != -1:
        cost = cost.replace("cents", "").strip()           
        cost = float(cost) * 0.01
    elif cost.find("$") != -1:
        cost = float(cost.replace("$", "").strip())     
    return cost
```

* Cleanse the column "Average Cost Per Person" using the function cleanse_cent() and Pandas's apply() function.

```
data1["Average Cost Per Person"] = data1["Average Cost Per Person"].apply(cleanse_cent)
data1
```
